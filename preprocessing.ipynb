```python
from collections import defaultdict, Counter

class NGramModel:
    def __init__(self, n=3):
        self.n = n
        self.ngrams = defaultdict(Counter)

    def train(self, lines):
        for line in lines:
            tokens = line.split()
            tokens = ['<s>'] * (self.n - 1) + tokens + ['</s>']
            for i in range(len(tokens) - self.n + 1):
                context = tuple(tokens[i:i+self.n-1])
                next_word = tokens[i+self.n-1]
                self.ngrams[context][next_word] += 1

    def predict(self, context, top_k=5):
        context = tuple(context.lower().split()[-(self.n-1):])
        candidates = self.ngrams.get(context, {})
        if not candidates:
            return []
        return [word for word, _ in candidates.most_common(top_k)]
```

```python
with open('DATA/preprocesseed.txt','r',encoding='utf-8') as f:
    lines=f.readlines()
# Train the model
model = NGramModel(n=3)
model.train(lines)
```

```python
# Try predicting
print(model.predict("ya allah"))
print(model.predict("aaj to"))
print(model.predict('waise to aaj '))
print(model.predict('waqy me ye'))
print(model.predict('mera to dil'))
print(model.predict('chalo aaj hum '))
print(model.predict('ji bilkul'))
print(model.predict('aap to'))
print(model.predict('mujhe pata nahi'))
print(model.predict('waise to chai bohot '))
```
